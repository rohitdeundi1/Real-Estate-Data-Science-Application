```python
df = df[df["price in crore"] != "Price on Request"]
```

```python
def convert_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == "Lac":
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df["price in crore"] = df["price in crore"].str.split(" ").apply(convert_price)
```

```python
df['price per sqft'] = df['price per sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
