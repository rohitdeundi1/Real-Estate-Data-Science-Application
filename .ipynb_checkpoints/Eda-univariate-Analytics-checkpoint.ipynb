## Society

```python
df['society'].head()
```

```python
df['society'].value_counts()
```

```python
df[df['society'] != 'independent']['society'].value_counts(normalize=True).cumsum().head(75)
```

```python
society_counts = df['society'].value_counts().values
# Frequency distribution for societies
frequency_bin = {
    "Very High (>100)" :(society_counts > 100).sum(),
    'High (50-100)' : ((society_counts >= 50) & (society_counts <=100)).sum(),
    'Average (10-49) ' : ((society_counts >= 10) & (society_counts < 50)).sum(),
    'Low (2-9)' : ((society_counts > 1) & (society_counts < 10)).sum(),
    'Very Low (1)' : (society_counts == 1).sum()
}
frequency_bin
```

```python
#top 10 socities
df[df['society'] != 'independent']['society'].value_counts().head(10).plot(kind = 'bar')
```

```python
df['society'].isnull().sum()
```

```python
df[df['society'].isnull()]
```

#### Observation
- Around 13% properties comes under independent tag
- There are 675 societies
- The top 75 societies have 50% of the properties and the rest 50% of the properties come under the remaining 600 societies
  - Very High (>100): Only 1 society has more than 100 listings.
  - High (50-100): 2 societies have between 50 to 100 listings.
  - Average (10-49) : 92 societies fall in this range with 10 to 49 listings each.
  - Low (2-9) : 273 societies have between 2 to 9 listings.
  - Very Low (1) : A significant number, 308 societies have only 1 listing.
- 1 missin value

## Sector

```python
# Unique sectors
df['sector'].value_counts().shape
```

```python
# Top 10 Societies
df['sector'].value_counts().head(10).plot(kind='bar')
```

```python
sector_counts = df['sector'].value_counts()
frequency_bin = {
    'Very High (>100)' : (sector_counts > 100).sum(),
    'High (50-100)' : ((sector_counts >= 50) & (sector_counts <= 100)).sum(),
    'Average (10-49)' : ((sector_counts >= 10) & (sector_counts < 50 )).sum(),
    'Low (2-9)' : ((sector_counts > 2) & (sector_counts > 9 )).sum(),
    'Very Low (1)' : (sector_counts == 1).sum()
}
frequency_bin
```

#### Observation
- There are total of 104 unique sectors in the database
- Frequency distrbution of sectors:
  - Very High (>100) : 3 sectors have more then 100 listings.
  - High (50-100) : 25 sectors have between 50 to 100 listings.
  - Average (10-49) : 60 sectors have between 10 to 49 listings.
  - Low (2-9) : 88 sectors have between 2 to 9 listings.
  - Very Low (1) : Intrestingly, there are no sectors with only 1 listing.
